```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd



```

```python
# Sample data
ages = [10,15,20,25,30,35,40,45,50,55,60,65,70,75,80]
genders = ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male']

# Create a Pandas DataFrame
df = pd.DataFrame({'Age': ages, 'Gender': genders})

# Create a histogram for the continuous variable (Age)
plt.hist(df['Age'], bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages')
plt.show()
```
